```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as pltcd 
from textblob import TextBlob
```

```python
df_path = "Bitcoin_tweets.csv"
df = pd.read_csv(df_path)

df.head(10)
```

```python
# remove nan value row in hastags column
df. dropna(subset = ["hashtags"], inplace=True)
df.head(5)
```

```python
# total value size
print("Tweet Size: ", len(df))
```

```python
# get only texts
df = df[['text', 'date', 'user_followers']]
df.columns = ['text', 'date', 'followers']
df
```

```python
# create a function to clean the tweets
def cleanTwt(twt):
    twt = re.sub("#bitcoin", 'bitcoin', twt) # removes the '#' from bitcoin
    twt = re.sub("#Bitcoin", 'Bitcoin', twt) # removes the '#' from Bitcoin
    twt = re.sub('#[A-Za-z0-9]+', '', twt) # removes any string with a '#'
    twt = re.sub('\\n', '', twt) # removes the '\n' string
    twt = re.sub('https:\/\/\S+', '', twt) # removes any hyperlinks
    return twt
```

```python
df['cleaned_tweets'] = df['text'].apply(cleanTwt)
df
```

```python
# create a function to get subjectivity
def getSubjectivity(twt):
    return TextBlob(twt).sentiment.subjectivity

# create a function to get the polarity
def getPolarity(twt):
    return TextBlob(twt).sentiment.polarity

# create two new columns called "Subjectivity" & "Polarity"
df['subjectivity'] = df['cleaned_tweets'].apply(getSubjectivity)
df['polarity'] = df['cleaned_tweets'].apply(getPolarity)
```

```python
df.head(5)
```

```python
# create a function get the sentiment text
def getSentiment(score):
    if score < 0:
        return "negative"
    elif score == 0:
        return "neutral"
    else:
        return "positive"
```

```python
# create a column to store the text sentiment
df['sentiment'] = df['polarity'].apply(getSentiment)

# show the data
df.head()
```

```python
df = df[['date', 'cleaned_tweets', 'followers']]
df.columns = ['time', 'tweets', 'followers']
df
```

```python
df = df[df['followers'] > 100000]
df
```

```python
df = df.sort_values(by = "time")
df
```

```python
df = df.reset_index(drop = True)
df
```
